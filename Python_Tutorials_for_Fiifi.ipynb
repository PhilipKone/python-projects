```python
import tkinter as tk

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, str(current) + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(tk.END, str(result))
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

root = tk.Tk()
root.title("Simple Calculator")

entry = tk.Entry(root, width=35, borderwidth=5)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+'
]

row_num = 1
col_num = 0
for button_text in buttons:
    if button_text == '=':
        btn = tk.Button(root, text=button_text, padx=40, pady=20, command=button_equal)
    elif button_text == '0':
         btn = tk.Button(root, text=button_text, padx=40, pady=20, command=lambda b=button_text: button_click(b))
    else:
        btn = tk.Button(root, text=button_text, padx=40, pady=20, command=lambda b=button_text: button_click(b))

    btn.grid(row=row_num, column=col_num)

    col_num += 1
    if col_num > 3:
        col_num = 0
        row_num += 1

clear_button = tk.Button(root, text="Clear", padx=100, pady=20, command=button_clear)
clear_button.grid(row=row_num, column=0, columnspan=4)


root.mainloop()
```

Now, let's create a simple calculator interface using `gradio`. We'll define a function that performs the calculation and then use `gradio.Interface` to create the web-based GUI.

```python
import gradio as gr

def calculator(num1, operator, num2):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "Cannot divide by zero"
        return num1 / num2
    else:
        return "Invalid operator"

interface = gr.Interface(
    fn=calculator,
    inputs=[gr.Number(label="Number 1"), gr.Radio(["+", "-", "*", "/"], label="Operator"), gr.Number(label="Number 2")],
    outputs="text",
    title="Simple Calculator"
)

interface.launch(share=True)
```
